```python
import datetime

# Lista para armazenar os registros
registros = []

# Função para registrar um novo processo
def registrar_processo():
    print("\n--- Registrar Processo ---")

    # Recebimento de material
    material = input("Material recebido: ")
    tipo_material = input("Tipo de material: ")
    data_recebimento = input("Data de recebimento (DD/MM/AAAA): ")
    hora_recebimento = input("Hora de recebimento (HH:MM): ")
    inspetor = input("Inspetor responsável: ")

    # Ordem de serviço
    ordem_servico = input("Ordem de serviço: ")
    setor = input("Setor: ")
    maquina = input("Máquina: ")
    ferramenteiro = input("Nome do ferramenteiro: ")

    # Horários do processo
    data_inicio = input("Data de início (DD/MM/AAAA): ")
    hora_inicio = input("Hora de início (HH:MM): ")
    hora_finalizada = input("Hora de finalização (HH:MM): ")

    # Inspeção e expedição
    inspecao_produto = input("Inspeção do produto (Aprovado/Reprovado): ")
    expedicao = input("Expedição (Sim/Não): ")

    # Criação do registro
    registro = {
        "Material": material,

        "Tipo Material": tipo_material,

        "Data Recebimento": data_recebimento,

        "Hora Recebimento": hora_recebimento,

        "Inspetor": inspetor,

        "Ordem Serviço": ordem_servico,

        "Setor": setor,

        "Máquina": maquina,

        "Ferramenteiro": ferramenteiro,

        "Data Início": data_inicio,

        "Hora Início": hora_inicio,

        "Hora Finalização": hora_finalizada,

        "Inspeção Produto": inspecao_produto,

        "Expedição": expedicao
    }

    registros.append(registro)
    print("\nRegistro adicionado com sucesso!")

# Função para exibir todos os registros
def exibir_registros():
    if not registros:
        print("\nNenhum registro encontrado.")
        return

    print("\n--- Registros ---")
    for i, registro in enumerate(registros, start=1):
        print(f"\nRegistro {i}:")
        for key, value in registro.items():
            print(f"{key}: {value}")

# Menu principal
def menu():
    while True:
        print("\n--- Menu ---")
        print("1. Registrar Processo")
        print("2. Exibir Registros")
        print("3. Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            registrar_processo()
        elif opcao == "2":
            exibir_registros()
        elif opcao == "3":
            print("Encerrando o sistema. Até logo!")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Executar o menu
if __name__ == "__main__":
    menu()
```

```python
import os
import pandas as pd
from datetime import datetime

# Configuração da pasta onde os dados serão salvos
PASTA_PROCESSOS = "processos"

if not os.path.exists(PASTA_PROCESSOS):
    os.makedirs(PASTA_PROCESSOS)

# Função para registrar um novo processo
def registrar_processo():
    dados = {
        "Recebimento de Material": input("Recebimento de Material: "),
        "Tipo de Material": input("Qual tipo de material: "),
        "Data Recebimento": input("Data (DD/MM/AAAA): "),
        "Hora Recebimento": input("Hora (HH:MM): "),
        "Inspetor": input("Inspetor: "),
        "Ordem de Serviço": input("Ordem de Serviço: "),
        "Setor": input("Setor: "),
        "Máquina": input("Máquina: "),
        "Nome do Ferramenteiro": input("Nome do Ferramenteiro: "),
        "Data Início": input("Data de Início (DD/MM/AAAA): "),
        "Hora Início": input("Hora de Início (HH:MM): "),
        "Hora Finalizada": input("Hora Finalizada (HH:MM): "),
        "Inspeção do Produto": input("Inspeção do Produto: "),
        "Expedição": input("Expedição (Sim/Não): ")
    }

    # Adicionando timestamp para nome único do arquivo
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    nome_arquivo = f"{PASTA_PROCESSOS}/processo_{timestamp}.csv"

    # Salvando dados como arquivo CSV
    pd.DataFrame([dados]).to_csv(nome_arquivo, index=False)
    print(f"Processo registrado e salvo como {nome_arquivo}")

# Função para gerar relatórios automáticos
def gerar_relatorio():
    arquivos = [f for f in os.listdir(PASTA_PROCESSOS) if f.endswith('.csv')]
    if not arquivos:
        print("Nenhum processo registrado para gerar relatório.")
        return

    processos = []
    for arquivo in arquivos:
        caminho = os.path.join(PASTA_PROCESSOS, arquivo)
        processos.append(pd.read_csv(caminho))

    # Concatenar todos os processos
    df_processos = pd.concat(processos, ignore_index=True)

    # Criar relatório consolidado
    relatorio_arquivo = f"{PASTA_PROCESSOS}/relatorio_consolidado.csv"
    df_processos.to_csv(relatorio_arquivo, index=False)
    print(f"Relatório consolidado salvo como {relatorio_arquivo}")

# Menu principal
def menu():
    while True:
        print("\n--- Sistema de Gerenciamento de Processos ---")
        print("1. Registrar novo processo")
        print("2. Gerar relatório")
        print("3. Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            registrar_processo()
        elif opcao == "2":
            gerar_relatorio()
        elif opcao == "3":
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Iniciar o sistema
menu()
```
